```python
import pandas as pd
```

```python
#https://archive.ics.uci.edu/ml/datasets/Electrical+Grid+Stability+Simulated+Data+#
```

```python
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/00471/Data_for_UCI_named.csv')
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
df.drop(columns=['stab'],inplace=True)
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df['stabf'] = encoder.fit_transform(df['stabf'])
df.head()
```

```python
X = df.drop(columns=['stabf'])
X.head()
```

```python
y = df['stabf']
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(X)
X_scalled = X.copy()
X_scalled  = pd.DataFrame(scaler.transform(X), columns=X.columns)
X_scalled.head()
```

```python
x_train_scalled = x_train.copy()
x_train_scalled  = pd.DataFrame(scaler.transform(x_train), columns=x_train.columns)
x_train_scalled.head()
```

```python
x_test_scalled = x_test.copy()
x_test_scalled  = pd.DataFrame(scaler.transform(x_test), columns=x_test.columns)
x_test_scalled.head()
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state= 1)
x_train_smote, y_train_smote = smote.fit_resample(x_train_scalled,y_train)
```

```python
from sklearn.metrics import recall_score, accuracy_score, precision_score, f1_score, confusion_matrix 
```

```python
#question 4
from xgboost import XGBClassifier
XGBclf = XGBClassifier(random_state=1)
XGBclf.fit(x_train, y_train)
XGBout = XGBclf.predict(x_test)
accuracy = accuracy_score(y_test,XGBout)
print( 'Accuracy:',(accuracy * 100))
```

```python
#question 6
from lightgbm import LGBMClassifier
LGBMclf = LGBMClassifier(random_state=1)
LGBMclf.fit(x_train, y_train)
LGBMout = LGBMclf.predict(x_test)
accuracy = accuracy_score(y_test,LGBMout)
print( 'Accuracy:',(accuracy * 100))
```
